## Bisection algorithm 


### Load Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
from sympy import *
```

### Bisection function

```python
def bisection(f,a,b,eps):
    i=1
    while abs(a-b)>eps:
        c = a+(b-a)/2
        if f(c)*f(a)<0:
            b = c
        else:
            a = c
        i+=1
    print('Number of iteration = {}'.format(i))
    print('The maximum of f(x) with respect to x is achieved at {}'.format(c))
```

### Figure 1

```python
x = np.linspace(1,5)
y = np.log(x)/(1+x)
plt.figure(figsize=(12,6))
plt.plot(x,y,linewidth=2)
plt.axvline(x=3.59112,linewidth=2,color='k', linestyle='--')
plt.ylabel('f(x)',weight='bold',fontsize=16)
plt.xlabel('x',weight='bold',fontsize=16)
plt.text(3.4, 0.02, r'$x* \approx 3.59112$', fontsize=16,rotation=90)
```

### f'(x) computation

```python
x = sym.Symbol('x')
fprime = sym.diff((sym.log(x))/(1+x))
f = lambdify(x, fprime, 'numpy')
fprime
```

#### Run the bisection function for initial values a=1, b=5 and convergence criteria=0.0001

```python
bisection(f,1,5,0.0001)
```
